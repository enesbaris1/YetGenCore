# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
import random
import time

kanallar=["Cartoon Network","Fox Tv","TLC","DMAX"]
tvbilgi={"Marka":"Vestel",
         "Model":"A03",
         "Renk":"Siyah"
         }


class kumanda():
    def __init__(self,tvbilgi,tvdurum,kanalsayisi,kanallar,ses):
        self.tvbilgi=tvbilgi
        self.tvdurum=tvdurum
        self.kanalsayisi=kanalsayisi
        self.kanallar=kanallar
        self.ses=ses
    
    def tvac(self):
        if self.tvdurum=="acik":
            print("Tv zaten acik!!")
        else:
            print("Tv acildi.")
            self.tvdurum="kapali"
    def tvkapat(self):
        if self.tvdurum=="kapali":
            print("Tv zaten kapali!!")
        else:
            print("Tv kapatildi.")
            self.tvdurum="kapali"
    def sesayar(self):
        while True:
            print("Mevcut ses durumunuz:",self.ses)
            control=input("\nSesi yukseltmek istiyorsaniz 1'e,\nKismak istiyorsaniz 2'ye,\nSes kismindan cikmak istiyorsaniz q harfine basiniz!")
            if control=="1":
                self.ses+=1
                print("Ses 1 kademe yukseltilmistir!")
            elif control=="2":
                self.ses-=1
                print("Ses 1 kademe kisilmistir!")
            elif control=="q":
                print("Ses kismindan cikiliyor...")
                time.sleep(2)
                break
            else:
                print("Hatali karakter girdiniz.Lutfen yeniden giris yapiniz!")
    def kanalekle(self):
        yeniKanal=input("Yeni eklenecek kanalin ismini giriniz:")
        self.kanallar.append(yeniKanal)
        self.kanalsayisi+=1
    def kanalsayiogren(self):
        print("Tv'de kayitli kanal sayisi.",self.kanalsayisi)
    def rastgele(self):
        rastgelekanal=random.choice(self.kanallar)
        print("Rastgele sectiginiz kanal:",rastgelekanal)
    def tvbilgisi(self):
        for i in self.tvbilgi.keys():
            print(i,":",self.tvbilgi[i])
    
def main():
    kumanda1=kumanda(tvbilgi,"kapali",len(kanallar),kanallar,10)
    print("*"*5,"KUMANDAYA HOSGELDİNİZ","*"*5)
    while True:
        print("Lutfen yapmak istediginiz islemin numarasini giriniz:\n1. Tv Aç\n2. Tv Kapat\n3. Ses Ayarlari\n4. Kanal Ekle\n5. Kanal Sayısını Öğrenme\n6. Rastgele Kanala Geçme\n7. Televizyon Bilgileri\nEger kumandayi kapatmak istiyorsaniz q tusuna basiniz!!")
        i1=input()
        if i1=="q":
            print("Kumandadan cikiliyor...")
            time.sleep(2)
            break
        elif i1=="1":
            kumanda1.tvac()
        elif i1=="2":
            kumanda1.tvkapat()
        elif i1=="3":
            kumanda1.sesayar()
        elif i1=="4":
            kumanda1.kanalekle()
        elif i1=="5":
            kumanda1.kanalsayiogren()
        elif i1=="6":
            kumanda1.rastgele()
        elif i1=="7":
            kumanda1.tvbilgisi()
        else:
            print("Yanlis bir sey girdiniz yeniden deneyiniz!!")

main()
```
