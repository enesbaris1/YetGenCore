# Problem 1 
Kullanıcıdan aldığı ürün bilgisini (ad, fiyat) urunler.txt dosyasına kayıt eden fonksiyonu yazınız.

````python
def kaydet(ad,fiyat):
    with open("fiyat.txt","a") as file:
        file.write(f"{ad}={fiyat}\n")

ad  = input("Lutfen girmek fiyat bilgisini girmek istediğiniz urunun ismini girin:")
fiyat = float(input("Lutfen Fiyat bilgisi giriniz:"))
kaydet(ad,fiyat)
````

# Problem 2
1.problemde oluşturulan text dosyası içerisinde kullanıcıdan alacağımız eski kelime isimli değişkene atadığımız 
ve değiştirilmesini istediğimiz kelimemizi yerini alıcak olan yeni kelime isimli değişkene atayın. Belirlenen kelimeleri text dosyasında değiştiren fonksiyonu yazınız.

````python
def buldegistir(eski,yeni):
    with open("fiyat.txt","r+") as file:
        icerik=file.read()
        file.seek(0)
        file.write(icerik.replace(eski,yeni))

eski=input("degistirmek istediginiz kelimeyi giriniz:")
yeni=input("yerine koyacaginiz kelime:")
buldegistir(eski,yeni)
````

# Problem 3
Vehicle sınıfının tüm değişkenlerini ve yöntemlerini devralacak bir alt sınıf olarak Bus sınıfı oluşturunuz.



```
Output
> Name: Vehicle sınıfının tüm değişkenlerini ve yöntemlerini devralacak bir alt sınıf olarak Bus sınıfı oluşturunuz.



```
Output
> Name: School Volvo Speed: 180 Mileage: 12
``` Speed: 180 Mileage: 12
```


````python
class Vehicle():
    def __init__(self, name, speed,mileage):
        self.name=name
        self.speed=speed
        self.mileage=mileage
    def showinfo(self):
        print(f"Name:{self.name}Speed:{self.speed}Mileage:{self.mileage}")

class Bus(Vehicle):
    pass

arac1 =Bus("School Volvo",180,12)
arac1.showinfo()
````
