# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
def vkibul(boy,kilo):
    vki = kilo/(boy**2)
    return vki

class kilolar():
    def __init__(self,isim,yas,boy,kilo,vki):
        self.isim = isim
        self.yas = yas
        self.boy = boy
        self.kilo = kilo
        self.vki = vki
class zayif(kilolar):
    pass
class normal(kilolar):
    pass
class fazla(kilolar):
    pass
class obez(kilolar):
    pass
class morbid(kilolar):
    pass

def bilgial(isim,yas,boy,kilo):
    isim=input("lütfen isminizi giriniz:")
    yas=int(input("lütfen yasinizi giriniz:"))
    boy=float(input("lütfen boyunuzu giriniz:"))
    kilo=float(input("lütfen kilonuzu giriniz:"))
    return isim,yas,boy,kilo

def main():
    with open("bilgiler.txt","w") as file:   
        i=int(input("kac kullanici girisi yapilacak:"))
        isim=""
        yas=0
        boy=1.0
        kilo=1.0
        while i>0:
            isim,yas,boy,kilo=bilgial(isim,yas,boy,kilo)
            vki=vkibul(boy,kilo)
            if vki<18.5:
                hasta = zayif(isim,yas,boy,kilo,vki)
                file.write(f"isim:{hasta.isim}\nyas:{hasta.yas}\nboy:{hasta.boy}\nkilo:{hasta.kilo}\nvki:{hasta.vki}-zayif")
            elif 18.5<=vki and vki<=24.9:
                hasta = normal(isim,yas,boy,kilo,vki)
                file.write(f"isim:{hasta.isim}\nyas:{hasta.yas}\nboy:{hasta.boy}\nkilo:{hasta.kilo}\nvki:{hasta.vki}-normal")
            elif 25<=vki and vki<=29.9:
                hasta = fazla(isim,yas,boy,kilo,vki)
                file.write(f"isim:{hasta.isim}\nyas:{hasta.yas}\nboy:{hasta.boy}\nkilo:{hasta.kilo}\nvki:{hasta.vki}-fazla")
            elif 30<=vki and vki<=39.9:
                hasta = obez(isim,yas,boy,kilo,vki)
                file.write(f"isim:{hasta.isim}\nyas:{hasta.yas}\nboy:{hasta.boy}\nkilo:{hasta.kilo}\nvki:{hasta.vki}-obez")
            else:
                hasta = morbid(isim,yas,boy,kilo,vki)
                file.write(f"isim:{hasta.isim}\nyas:{hasta.yas}\nboy:{hasta.boy}\nkilo:{hasta.kilo}\nvki:{hasta.vki}-morbid")
            i-=1
main()
    
            

```

# Problem 2
**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
menu = {"elma":1,
        "armut":3,
        "cilek":4,
        "seftali":9,
        "mango":10}
def menugoster(menu):
    for i in menu.keys():
        print(f"{i + ':':<10} {menu[i]}")

def main():
    toplam=0
    while True:
        giris=int(input("Eger menuyu gormek istiyorsaniz 1'e,\nSatin alim yapmak istiyorsaniz 2'ye,\nCikis yapmak istiyorsaniz 0'a basin lutfen."))
        if giris==0:
            break
        elif giris==1:
            menugoster(menu)
        elif giris==2:
            giris2=int(input("kac farkli urun almak istersiniz:"))
            while giris2 > 0:
                urun = input("hangi urunu almak istersiniz:")
                urun = urun.lower()
                if urun in menu.keys():
                    adet = int(input("bu urunden kac adet almak istersiniz:"))
                    toplam += menu[urun]*adet
                    giris2-=1
                else:
                    print("yanlis giris yaptiniz!\nBir daha deneyin!")
    
    print("Smoothie'niz hazir!\nOdemeniz gereken tutar:",toplam)

main()
```
